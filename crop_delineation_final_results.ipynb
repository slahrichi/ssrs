```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./crop_delineation_batch_results_ORIGINAL.csv', index_col=0)
df2 = pd.read_csv('./crop_delineation_batch_results_c2.csv', index_col=0)
```

```python
df_pivot = pd.pivot_table(df, values = 'iou', index='datasize', columns='encoder')
df_pivot_c2 = pd.pivot_table(df2, values = 'iou', index='datasize', columns='encoder')
```

```python
plt.plot(df_pivot.index, df_pivot['supervised'], label='supervised|imagenet')
plt.plot(df_pivot.index, df_pivot['swav-imagenet'], label='swav|imagenet')
plt.plot(df_pivot.index, df_pivot['swav-c1'], label='swav|imagenet|target')
plt.plot(df_pivot_c2.index, df_pivot_c2['swav-c2'], label='swav|target')
plt.xscale('log')
plt.xlabel("Datasize")
plt.ylabel("Pixel IoU")
plt.legend()
```
